```python
import numpy as np
import pandas as pd
```

```python
column_names= ['user_id', 'item_id', 'rating', 'timestamp']
df=pd.read_csv('users.data', sep= '\t', names=column_names)
```

```python
df.head()
```

```python
#datasetin uzunluğu (kaç kayıt var?)
len(df)
```

```python
filmler=pd.read_csv("movie_id_titles.csv")
filmler.head()
```

```python
len(filmler)
```

```python
#iki verisetini birleştirelim
df=pd.merge(df,filmler, on="item_id")
df.head()
```

# #Recommendation Sistemi Kuruyoruz

```python
film = df.pivot_table(index='user_id', columns='title', values='rating')
film.head()
```

```python
type (film)
```

```python
#amaç:starwars filmine benzer film önermek

starwars_ratings=film['Star Wars (1977)']
starwars_ratings.head()
```

```python
#corrwith() metodu ile star wars filmi ile kolerasyon hesaplatacağız.

sw_benzeri= film.corrwith(starwars_ratings)
```

```python
sw_benzeri
```

```python
type(sw_benzeri)
```

```python
#Nan kayıtları temizleyeceğiz.
duzeltme_sw = pd.DataFrame(sw_benzeri, columns = ['Correlation'])
duzeltme_sw.dropna(inplace = True) #dropna herhangi bir satırda NaN değeri varsa onu siler
duzeltme_sw.head()
```

```python
#elde ettiğimiz dataframe'i sıralayarak starwars'a en yakın tavsiye
#edilecek filmi göreceğiz

duzeltme_sw.sort_values('Correlation', ascending=False).head(10)
```

Burada alakasız sonuçlar çıktı bunun nedeni bu filmlerin çok az oy alması. Bu durumu düzeltmek için 100'den az oy alan filmleri ekleyeceğiz.
Bu amaçla ratings isimli bir dataframe oluşturup buradca her filmin kaç oy aldığını tutacağız.

```python
df.head()
```

```python
df.drop(['timestamp'],axis = 1)
```

```python
#her filmin ortalama rating değerini bulacağız (mean value)
ratings = pd.DataFrame(df.groupby('title') ['rating'].mean())
```

```python
#büyükten küçüğe sıralama
ratings.sort_values('rating', ascending=False).head()
```

```python
#şimdi her filmin aldığı oy sayısına bakacağız
ratings ['rating_oy_sayisi']=pd.DataFrame(df.groupby('title')['rating'].count())
```

```python
ratings.head()
```

```python
#en çok oy alan filmleri büyükten küçüğe doğru sıralayıp bakacağız.
ratings.sort_values('rating_oy_sayisi', ascending=False).head()
```

```python
#esas amacımıza dönelim ve corr_starwars dataframe'imize rating_oy_sayisi sütununu ekleyelim (join ile) 
```

```python
duzeltme_sw.sort_values('Correlation', ascending=False).head(10)
```

```python
duzeltme_sw=duzeltme_sw.join(ratings['rating_oy_sayisi'])
duzeltme_sw.head()
```

```python
duzeltme_sw[duzeltme_sw['rating_oy_sayisi']>100].sort_values('Correlation',ascending=False).head()
```
